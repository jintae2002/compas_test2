```python
# 경찰서명 통일하는 함수
def make_police_name(row) :
    if '서울' in row:
        row2 = row[:-1] + '경찰서'
        return row2
    elif '경기' in row:
        row2 = row[4:8] + '경찰서'
        return row2
    elif '세종' in row:
        row2 = '세종경찰서'
        return row2
    elif '경남' in row :
        if '창원' in row :
            row2 = row[2:6] + '경찰서'
            return row2
        elif '마산' in row:
            row2 = row[2:6] + '경찰서'
            return row2
        else :
            row2 = row[2:4] + '경찰서'
            return row2
```

```python
df_f1['medium_category'] = \
df_f1['medium_category'].apply(make_police_name)
```

```python
col_list = ['대분류','중분류','종합체감안전도점수','분야별안전도','범죄안전도','절도폭력안전도','강도살인안전도','교통사고안전도','법질서안전도','전반적안전도']
df_f1.columns = col_list
df_f1
```

```python
def make_integrated_df(df, category) :
    dff = pd.DataFrame(index=df['중분류'].unique())
    
    cond1 = df[df['대분류']=='2017 상반기(경찰서별)'][category].values
    dff['2017상'] = cond1 
    
    cond2 = df[df['대분류']=='2017 하반기(경찰서별)'][category].values
    dff['2017하'] = cond2 

    cond3 = df[df['대분류']=='2018 상반기(경찰서별)'][category].values
    dff['2018상'] = cond3
    
    cond4 = df[df['대분류']=='2018 하반기(경찰서별)'][category].values
    dff['2018하'] = cond4 
    
    cond5 = df[df['대분류']=='2019 상반기(경찰서별)'][category].values
    dff['2019상'] = cond5 

    cond6 = df[df['대분류']=='2019 하반기(경찰서별)'][category].values
    dff['2019하'] = cond6
    
    dff = dff.stack().to_frame().reset_index().rename(columns={'level_0':'경찰서명','level_1':'year',0:category})
    return dff
```

```python
df_rob = make_integrated_df(df_f1,'절도폭력안전도')
df_rob
```
