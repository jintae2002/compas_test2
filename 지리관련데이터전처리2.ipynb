- 년도만 있는것을 상/하반기 나눌것 
- 2017년 이전은 2017상반기로 땡김

```python
import pandas as pd
import numpy as np
```

# 보안등현황

```python
df9_light = pd.read_csv('경찰서별_년도별_보안등현황.csv')
df9_light
```

```python
df9_light['2017상'] = df9_light['2017년']
df9_light['2017하'] = np.ceil((df9_light['2017년'] + df9_light['2018년'])/2)
df9_light['2017하'] = df9_light['2017하'].astype('int')

df9_light['2018상'] = df9_light['2018년']
df9_light['2018하'] = np.ceil((df9_light['2018년'] + df9_light['2019년'])/2)
df9_light['2018하'] = df9_light['2018하'].astype('int')

df9_light['2019상'] = df9_light['2019년']
df9_light['2019하'] = np.ceil((df9_light['2019년'] + df9_light['2020년'])/2)
df9_light['2019하'] = df9_light['2019하'].astype('int')

df9_light['2020상'] = df9_light['2020년']
df9_light['2020하'] = np.ceil((df9_light['2020년'] + df9_light['2021년'])/2)
df9_light['2020하'] = df9_light['2020하'].astype('int')

df9_light['2021상'] = df9_light['2021년']
```

```python
df9_light_1 = df9_light[['경찰서별', '2017상','2017하','2018상','2018하','2019상','2019하','2020상','2020하','2021상']]
df9_light_1
```

```python
df9_light_final = \
df9_light_1.set_index('경찰서별').stack().to_frame().reset_index().rename(columns={'경찰서별':'경찰서명','level_1':'year',0:'보안등수'})
```

```python
df9_light_final.to_csv('전처리_9.보안등현황.csv',encoding='utf-8-sig')
```
