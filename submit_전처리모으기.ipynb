## 설문결과데이터


```python
res = pd.read_csv('4.체감안전도_통계(2017~2019).csv')
res['large_category'] = res.large_category.str[:6]
res = res[~(res.medium_category.str.contains('\d'))]
res = res[~(res.medium_category.isin(['남자','여자']))]
res = res[~(res.medium_category.isin(['남성','여성']))]
res.medium_category = res.medium_category.str.replace('서$', '')
res.medium_category = res.medium_category + '경찰서'
res.medium_category = res.medium_category.str.replace( '경기남부','')
res.medium_category = res.medium_category.str.replace( '경남','')
res.medium_category = res.medium_category.str.replace( '충남','')
res.rename(columns={'large_category':'year', 'medium_category':'police'}, inplace=True)
res = res[['year', 'police',' rob_satety_q1', 'mur_safety_q2', 'ta_safety_q3', 'raw_odder_q4','overall_q5']]
res.year = res.year.str.replace(' ', '')
res.reset_index(inplace=True, drop=True)
res
```

```python
res.police.nunique(), res.police.unique()
```

```python
res.year.unique()
```

```python
res.police = res.police.str.replace( '서울강경찰서','서울강서경찰서')
res.police = res.police.str.replace( '서울수경찰서','서울수서경찰서')
```

```python
res_pol = res.copy()
res_pol
```
