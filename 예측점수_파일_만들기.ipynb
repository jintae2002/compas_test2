```python
df_final = pd.DataFrame()
df_final['순번'] = df['순번']
df_final['경찰서명'] = df['경찰서명']
df_final['경찰서명'] = df_final['경찰서명'].apply(lambda x : x.strip())
df_final
```

```python
li1 = []
for i in df_final['경찰서명'] :
    li1.append(i)
print(li1)
```

```python
df_final = df_final.merge(df1, how='left', on='경찰서명')
df_final
```

```python
df_final = df_final.merge(df2, how='left', on='경찰서명')
df_final
```

```python
df3 = df3.iloc[:,1:]
df3.columns = ['경찰서명','문항3. 교통사고 안전도\n(문항3 점수*10)']
df3
```

```python
df_final = df_final.merge(df3, how='left', on='경찰서명')
df_final
```

```python
df4 = df4.iloc[:,1:]
df4.columns = ['경찰서명','문항4. 법질서 준수도\n(문항 4점수 * 10)']
df4
```

```python
df_final = df_final.merge(df4, how='left', on='경찰서명')
df_final
```

```python
df5.columns = ['경찰서명','문항5. 전반적 안전도\n(문항 5점수 * 10)']
df5
```

```python
df_final = df_final.merge(df5, how='left', on='경찰서명')
df_final
```

```python
df_final['범죄 안전도\n(절도폭력안전도+강도살인안전도)/2'] = 0.5*df_final['문항1. 절도 폭력안전도(문항1 점수 * 10)'] + 0.5*df_final['문항2. 강도 살인 안전도(문항2 점수 * 10)']
df_final
```

```python
df_final['분야별 안전도\n(범죄안전도*0.343) + \n(교통사고안전도*0.305)\n + (법질서준수도*0.352)'] = \
0.343*df_final['범죄 안전도\n(절도폭력안전도+강도살인안전도)/2'] + 0.305*df_final['문항3. 교통사고 안전도\n(문항3 점수*10)'] + \
0.352*df_final['문항4. 법질서 준수도\n(문항 4점수 * 10)']
df_final
```

```python
df_final['종합체감안전도_점수 \n(전반적안전도*0.3) + (분야별안전도*0.7)'] = \
0.3*df_final['문항5. 전반적 안전도\n(문항 5점수 * 10)'] + 0.7*df_final['분야별 안전도\n(범죄안전도*0.343) + \n(교통사고안전도*0.305)\n + (법질서준수도*0.352)']
df_final
```

```python
df_final2 = df_final[['순번','경찰서명','종합체감안전도_점수 \n(전반적안전도*0.3) + (분야별안전도*0.7)','분야별 안전도\n(범죄안전도*0.343) + \n(교통사고안전도*0.305)\n + (법질서준수도*0.352)',
                      '범죄 안전도\n(절도폭력안전도+강도살인안전도)/2','문항1. 절도 폭력안전도(문항1 점수 * 10)','문항2. 강도 살인 안전도(문항2 점수 * 10)',
                      '문항3. 교통사고 안전도\n(문항3 점수*10)','문항4. 법질서 준수도\n(문항 4점수 * 10)','문항5. 전반적 안전도\n(문항 5점수 * 10)']]
df_final2
```
